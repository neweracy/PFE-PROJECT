#### GETTING THE REQUIRED DATA FROM 'data.csv' file


```python
# assigning dataset to the information in ./data.csv file

dataset = pd.read_csv('./data.csv')
```

#### CLEANING UP THE DATA ( REMOVING ALL NULL VALUES AND DUPLICATES )
---

```python
# This removes all the spaces between words in the columns and replaces them with underscores '_'
dataset.columns = dataset.columns.str.replace(' ', '_')
```

```python
dataset
```

```python
# checking the information about the data
dataset.info()
```

<p>
  The isnull() method is used to check whether a value in a pandas DataFrame or Series is missing or not. This method returns a boolean mask indicating where the values are null (True) or not null (False).
</p>

```python
# checking for null values
dataset.isnull().sum()
```

```python
# this code is to find the mean of each of the columns
dataset_mean_data = dataset.mean()
```

```python
# this variable represents the mean of each column of the dataset
dataset_mean_data
```

#### REPLACING ALL NULL VALUES WITH THE MEAN OF EACH COLUMN AND REMOVING ALL DUPLICATES

<p>We can fill null values with the mean, median, or mode of the respective column using the fillna() method. This is a common approach and is useful when there are only a few null values in the dataset just like this dataset which we will be using the mean value and the drop_duplicate() method drop all duplicates in the dataset.</p>

```python
dataset = dataset.fillna(dataset_mean_data).drop_duplicates()
```

### ASSIGNING THE MEAN OF EACH COLUMN TO VARIABLES

```python
meanFA = dataset.fixed_acidity.mean()
meanVA = dataset.volatile_acidity.mean()
meanCA = dataset.citric_acid.mean()
meanRS = dataset.residual_sugar.mean()
meanC = dataset.chlorides.mean()
meanFSD = dataset.free_sulfur_dioxide.mean()
meanTSD = dataset.total_sulfur_dioxide.mean()
meanD = dataset.density.mean()
meanPH = dataset.pH.mean()
meanS = dataset.sulphates.mean()
meanA = dataset.alcohol.mean()
meanQ = dataset.quality.mean()
indexing = dataset.index
```

#### THE MEAN FOR EACH COLUMN

```python
print(f'The mean value of fixed_acidity  is {meanFA}')
print(f'The mean value of volatile_acidity is {meanVA}')
print(f'The mean value of citric_acid is {meanCA}')
print(f'The mean value of residual_sugar is {meanRS}')
print(f'The mean value of chlorides is {meanC}')
print(f'The mean value of free_sulfur_dioxide is {meanFSD}')
print(f'The mean value of total_sulfur_dioxide is {meanTSD}')
print(f'The mean value of density is {meanD}')
print(f'The mean value of pH is {meanPH}')
print(f'The mean value of sulphate is {meanS}')
print(f'The mean value of alcohol is {meanA}')
print(f'The mean value of quality is {meanQ}')
```

## GRAPHING THE DATA
---

We can see some hidden inferences visually through the graphs. The Bar graphs, Pie charts, and histograms show the distributions , outliner and relationships among variables in the dataset

```python
# data for pie chart
# total columns excluding ID is 12
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple',
          'pink', 'brown', 'gray', 'black', 'violet', 'beige']
explode = (0.5, 0.5, 1, 1, 1, 0, 0, 0.5, 0, 0.5, 0, 0.5)
mean_values = [meanFA, meanVA, meanCA, meanRS, meanC,
              meanFSD, meanTSD, meanD, meanPH, meanS, meanA, meanQ]
mean_topics = ['FA',
              'VA',
              'CA',
              'RS',
              'C',
              'FSD',
              'TSD',
              'density',
              'pH',
              'sulphates',
              'alcohol',
              'quality']
```

```python
x_title = 'Column names'
y_title = 'Mean Values'
legendLabels = ['fixed_acidity (FA)',
                'volatile_acidity (VA)',
                'citric_acid (CA)',
                'residual_sugar (RS)',
                'chlorides (C)',
                'free_sulfur_dioxide (FSD)',
                'total_sulfur_dioxide (TSD)',
                'density',
                'pH',
                'sulphates',
                'alcohol',
                'quality']
```

```python
labels = mean_topics
sizes = mean_values
```

#### COMPARING THE MEAN VALUES OF THE DATA USING A PIE CHART

```python
# plotting the pie chart
plt.pie(sizes, labels=labels, explode=explode,
        autopct='%1.1f%%', pctdistance=1.01, shadow=True, colors=colors, labeldistance=1.2)
plt.axis('equal')
plt.legend(title='index', labels=legendLabels, loc='upper left',)
plt.title('Mean Values for each of the columns in a pie chart')
plt.figure(figsize=(15, 18))
plt.show()
```

#### COMPARING THE MEAN VALUES OF THE DATA USING A BAR GRAPH

```python
# plotting a bar graph

fig = plt.figure(figsize=(10, 5))

plt.bar(labels, sizes, width=0.5, color=colors)
plt.xlabel(x_title)
plt.ylabel(y_title)
plt.title('Comparing mean of columns in a bar chart')
plt.legend(loc='upper right', )
plt.show()
```
